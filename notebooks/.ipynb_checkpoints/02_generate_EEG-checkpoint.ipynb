```python
import os
import torch
import cv2
import pandas as pd
import numpy as np
from glob import glob
from tqdm import tqdm
import joblib

import random

import matplotlib.pyplot as plt
```

```python
print("PyTorch:", torch.__version__)
print("OpenCV:", cv2.__version__)
```

```python
class CFG:
    verbose = 1  # Verbosity
    seed = 35  # Random seed
    preset = "efficientnetv2_b2_imagenet"  # Name of pretrained classifier
    image_size = [400, 300]  # Input image size
    epochs = 13 # Training epochs
    batch_size = 64  # Batch size
    lr_mode = "cos" # LR scheduler mode from one of "cos", "step", "exp"
    drop_remainder = True  # Drop incomplete batches
    num_classes = 6 # Number of classes in the dataset
    fold = 0 # Which fold to set as validation data
    class_names = ['Seizure', 'LPD', 'GPD', 'LRDA','GRDA', 'Other']
    label2name = dict(enumerate(class_names))
    name2label = {v:k for k, v in label2name.items()}
```

```python
torch.manual_seed(CFG.seed)
random.seed(CFG.seed)
np.random.seed(CFG.seed)
```

```python
BASE_PATH = "../data"

SPEC_DIR = "./tmp/dataset/generated-eegs"

os.makedirs(SPEC_DIR + '/train_eegs', exist_ok=True)
os.makedirs(SPEC_DIR + '/test_eegs', exist_ok=True)
```

```python
# Train + Valid
df = pd.read_csv(f'{BASE_PATH}/train.csv')

df['eeg_path'] = f'{BASE_PATH}/train_eegs/'+df['eeg_id'].astype(str)+'.parquet'
df['spec_path'] = f'{BASE_PATH}/train_spectrograms/'+df['spectrogram_id'].astype(str)+'.parquet'
df['spec2_path'] = f'{SPEC_DIR}/train_spectrograms/'+df['spectrogram_id'].astype(str)+'.npy'
df['class_name'] = df.expert_consensus.copy()
df['class_label'] = df.expert_consensus.map(CFG.name2label)
display(df.head(10))

# Test
test_df = pd.read_csv(f'{BASE_PATH}/test.csv')
test_df['eeg_path'] = f'{BASE_PATH}/test_eegs/'+test_df['eeg_id'].astype(str)+'.parquet'
test_df['spec_path'] = f'{BASE_PATH}/test_spectrograms/'+test_df['spectrogram_id'].astype(str)+'.parquet'
test_df['spec2_path'] = f'{SPEC_DIR}/test_spectrograms/'+test_df['spectrogram_id'].astype(str)+'.npy'
display(test_df.head(2))
```

## Process EEGs from parquet to .npy

```python
def process_eeg(eeg_id, split="train"):
    eeg_path = f"{BASE_PATH}/{split}_eegs/{eeg_id}.parquet"
    eeg_data = pd.read_parquet(eeg_path)
    
    # Fill NaN values with 0
    eeg_data = eeg_data.fillna(0).values

    eeg_data = eeg_data.astype("float32")

    np.save(f"{SPEC_DIR}/{split}_eegs/{eeg_id}.npy", eeg_data)

# Get unique eeg_ids of train and valid data
eeg_ids = df["eeg_id"].unique()

# Parallelize the processing using joblib for training data with a single tqdm progress bar
for _ in tqdm(Parallel(n_jobs=-1, backend="loky")(
    delayed(process_eeg)(eeg_id, "train") for eeg_id in eeg_ids
), total=len(eeg_ids)):
    pass

# Get unique eeg_ids of test data
test_eeg_ids = test_df["eeg_id"].unique()

# Parallelize the processing using joblib for test data with a single tqdm progress bar
for _ in tqdm(Parallel(n_jobs=-1, backend="loky")(
    delayed(process_eeg)(test_eeg_id, "test") for test_eeg_id in test_eeg_ids
), total=len(test_eeg_ids)):
    pass
```

## Simple, Individual Channel Spectrogram Generation Using STFT

```python
def generate_spectrogram(eeg_npy_path, save_dir, nperseg=256, noverlap=128):
    eeg_data = np.load(eeg_npy_path)

    os.makedirs(save_dir, exist_ok=True)
    
    # Generate spectrograms for each channel
    for i, channel_data in enumerate(eeg_data.T):  # Transpose to iterate over channels
        f, t, Zxx = stft(channel_data, nperseg=nperseg, noverlap=noverlap)
        
        # Compute the magnitude (power) of the spectrogram
        spectrogram = np.abs(Zxx)
        
        # Normalize and log scale the spectrogram
        spectrogram = np.log1p(spectrogram)
        
        # Save the spectrogram
        channel_dir = os.path.join(save_dir, f"channel_{i}")
        os.makedirs(channel_dir, exist_ok=True)
        spectrogram_path = os.path.join(channel_dir, os.path.basename(eeg_npy_path))
        np.save(spectrogram_path, spectrogram)
```

```python
train_eeg_dir = './tmp/dataset/generated-eegs/train_eegs'
test_eeg_dir = './tmp/dataset/generated-eegs/test_eegs'
save_spectrogram_train_dir = './tmp/dataset/generated-spectrograms/train_spectrograms'
save_spectrogram_test_dir = './tmp/dataset/generated-spectrograms/test_spectrograms'
os.makedirs(save_spectrogram_train_dir, exist_ok=True)
os.makedirs(save_spectrogram_test_dir, exist_ok=True)
```

```python
from glob import glob
train_eeg_files = glob(os.path.join(train_eeg_dir, '*.npy'))
test_eeg_files = glob(os.path.join(test_eeg_dir, '*.npy'))
```

```python
# confirm channels
eeg_data = np.load(os.path.join(train_eeg_files[0]))
eeg_data.shape
```

```python
from joblib import Parallel, delayed
from tqdm import tqdm

def process_spectrogram(file_path, save_dir):
    generate_spectrogram(file_path, save_dir)

def parallel_process(files, save_dir, desc):
    Parallel(n_jobs=-1, backend="loky")(
        delayed(process_spectrogram)(file_path, save_dir)
        for file_path in tqdm(files, desc=desc)
    )

parallel_process(train_eeg_files, save_spectrogram_train_dir, "Processing Train EEGs")
parallel_process(test_eeg_files, save_spectrogram_test_dir, "Processing Test EEGs")
```

```python
def plot_spectrogram(spectrogram_file):
    spectrogram = np.load(spectrogram_file)
    
    plt.figure(figsize=(10, 6))
    plt.imshow(spectrogram, aspect='auto', origin='lower')
    plt.title('Spectrogram')
    plt.ylabel('Frequency')
    plt.xlabel('Time')
    plt.colorbar(label='Log Amplitude')
    plt.show()
```

## EEG to MelSpectrogram

```python
import librosa
import librosa.display
```

```python
def generate_mel_spectrogram(eeg_npy_path, save_dir, nperseg=256, noverlap=128, n_mels=128, sr=100):
    eeg_data = np.load(eeg_npy_path)
    os.makedirs(save_dir, exist_ok=True)

    mel_spectrograms = []
    
    # Generate mel-spectrograms for each channel
    for channel_data in eeg_data.T:  # Transpose to iterate over channels
        f, t, Zxx = stft(channel_data, nperseg=nperseg, noverlap=noverlap)
        
        # Convert to magnitude spectrogram
        spectrogram = np.abs(Zxx)
        
        # Convert the spectrogram to mel scale
        mel_spectrogram = librosa.feature.melspectrogram(S=spectrogram, sr=sr, n_mels=n_mels)
        
        # Convert to log scale (dB)
        mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)
        
        # Append the mel-spectrogram to the list
        mel_spectrograms.append(mel_spectrogram_db)

    combined_mel_spectrogram = np.concatenate(mel_spectrograms, axis=0)
    spectrogram_path = os.path.join(save_dir, os.path.basename(eeg_npy_path))
    np.save(spectrogram_path, combined_mel_spectrogram)
```

```python
mel_spectrogram_train_dir = './tmp/dataset/mel-spectrograms/train_spectrograms'
mel_spectrogram_test_dir = './tmp/dataset/mel-spectrograms/test_spectrograms'
os.makedirs(mel_spectrogram_train_dir, exist_ok=True)
os.makedirs(mel_spectrogram_test_dir, exist_ok=True)
```

```python
def process_mel_spectrograms(files, save_dir, desc):
    # Run the process in parallel without tqdm inside the parallelized function
    with tqdm(total=len(files), desc=desc) as pbar:
        results = Parallel(n_jobs=-1, backend="loky")(
            delayed(generate_mel_spectrogram)(file_path, save_dir)
            for file_path in files
            
        )
        pbar.update(len(files))
```

```python
process_mel_spectrograms(train_eeg_files, mel_spectrogram_train_dir, "Processing Train Mel-Spectrograms")
process_mel_spectrograms(test_eeg_files, mel_spectrogram_test_dir, "Processing Test Mel-Spectrograms")
```

```python
def visualize_mel_spectrogram(mel_spectrogram_file, sr=100, n_mels=128):
    mel_spectrogram = np.load(mel_spectrogram_file)

    plt.figure(figsize=(10, 6))
    librosa.display.specshow(mel_spectrogram, sr=sr, x_axis='time', y_axis='mel', cmap='viridis')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel-Spectrogram')
    plt.ylabel('Mel Frequency')
    plt.xlabel('Time')
    plt.show()
```

```python
sample_mel_spectrogram_file = os.path.join(mel_spectrogram_train_dir, '582999.npy')
visualize_mel_spectrogram(sample_mel_spectrogram_file)
```
